```python
# most sold country
country_count=df['Country'].value_counts()
max_soled_country=country_count.nlargest(1)
country=max_soled_country.index[0]
max_no_soled=max_soled_country.values[0]
print(country)
print(max_no_soled)
```

```python
#most sold product
product_count=df[' Product '].value_counts()
max_soled_product=product_count.nlargest(1)
product=max_soled_product.index[0]
max_no_soled=max_soled_product.values[0]
print(product)
print(max_no_soled)
```

```python
#most profit product
most_profit_product=df.sort_values(' Profit ',ascending=False).iloc[0]
print(most_profit_product[' Product '])
```

```python
#product with high sales price
most_salesprice_product=df.sort_values(' Sale Price ',ascending=False).iloc[0]
print(most_salesprice_product[' Product '])
```

```python
#product with highest gross revenue
high_grossrevenue_product=df.sort_values(' Gross Sales ',ascending=False).iloc[0]
print(high_grossrevenue_product[' Product '])
```

```python
sns.scatterplot(x=' Sale Price ',y=' Manufacturing Price ',hue=' Product ',data=df)
plt.show()
```

```python
y=(df['Units Sold'])
x=(df[' Product '])
plt.bar(x,y)
plt.xlabel('product')
plt.ylabel('unit sold')
plt.title('TOTAL PRODUCTS SOLD')
plt.show()
```

```python
SALES_BY_YEAR=df['Year'].value_counts()
sns.barplot(x=SALES_BY_YEAR.index,y=SALES_BY_YEAR.values)
plt.xlabel('YEAR')
plt.ylabel('COUNT OF SALES')
plt.title('SALES COUNT BY YEAR ')
plt.show()
```

```python
selling_of_product=df[' Product '].value_counts()
sns.barplot(x=selling_of_product.index,y=selling_of_product.values)
plt.xlabel('product')
plt.ylabel('count of product sold')
plt.title('sold product count')
plt.show()
```

```python
sns.scatterplot(x='  Sales ',y=' COGS ',hue=' Product ',data=df)
plt.title('SALES AND COGS WITH PRODUCT')
plt.show()
```

```python
sns.scatterplot(x=' Gross Sales ',y='  Sales ',hue=' Product ',data=df)
plt.title('GROSS SALES & SALES WITH PRODUCT')
plt.show()
```

```python

plt.boxplot(df['Year'].value_counts(),patch_artist=True,labels=['  Sales '])
plt.show()
```

```python
numerical_columns = df.select_dtypes(include='number')
numerical_columns 

correlation_matrix = numerical_columns.corr()
correlation_matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap Correlation Plot')
plt.show()
```

```python
sales_per_segment = df.groupby('Segment')['  Sales '].sum()

plt.figure(figsize=(8, 8))
plt.pie(sales_per_segment, labels=sales_per_segment.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Sales per segment')
plt.show()
```

```python
plt.hist(df['  Sales '],bins=10,edgecolor='black')

plt.xlabel('sales')

plt.ylabel('variation')

plt.title('plot of sales range')

plt.show()
```
